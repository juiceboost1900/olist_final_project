```python
# Single-cell export of all Silver tables to OneLake Files (CSV)
# Location: Lakehouse -> Files/exports/silver_csv/

tables = [
    "sl_orders",
    "sl_order_items",
    "sl_order_reviews",
    "sl_sellers",
    "sl_customers",
    "sl_geolocation",
    "sl_products",
    "sl_product_category_translation",
    "sl_payments",
]

base_path = "Files/exports/silver_csv"

for t in tables:
    df = spark.table(f"dbo.{t}")
    out_path = f"{base_path}/{t}"
    
    (df.coalesce(1)               # one CSV file per table (submission-friendly)
       .write
       .mode("overwrite")
       .option("header", True)
       .csv(out_path))
    
    print(f"Exported {t} -> {out_path}")

print("All Silver tables exported successfully.")
```
