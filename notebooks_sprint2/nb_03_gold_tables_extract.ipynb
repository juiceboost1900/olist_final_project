```python
# Export Gold tables to OneLake Files as CSV (one folder per table)
# Location: Lakehouse -> Files/exports/gold_csv/<table_name>/

tables = [
    "dim_date",
    "dim_customer",
    "dim_product",
    "dim_seller",
    "dim_order",
    "dim_payment",
    "fact_order_items",
    "fact_seller_daily_sales",
]

base_path = "Files/exports/gold_csv"

for t in tables:
    df = spark.table(f"dbo.{t}")  
    out_path = f"{base_path}/{t}"

    (df.coalesce(1)                
       .write
       .mode("overwrite")
       .option("header", True)
       .csv(out_path))

    print(f"Exported {t} -> {out_path}")

print("All Gold tables exported successfully.")
```
